```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

```python
df = pd.read_csv("../data/raw/amazon.csv")

df.head()
```

```python
print("Shape:", df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df["discounted_price"] = (
    df["discounted_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df["discounted_price"].head()
```

```python
df["actual_price"] = (
    df["actual_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df["actual_price"].head()
```

```python
df["discount_percentage"] = (
    df["discount_percentage"]
    .str.replace("%", "", regex=False)
    .astype(float)
)

df["discount_percentage"].head()
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df["rating"].head()
```

```python
df["rating_count"] = (
    df["rating_count"]
    .str.replace(",", "", regex=False)
)

df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")

df["rating_count"].head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# Fill missing rating with median
df["rating"] = df["rating"].fillna(df["rating"].median())

# Fill missing rating_count with median
df["rating_count"] = df["rating_count"].fillna(df["rating_count"].median())

df.isnull().sum()
```

```python
df = df.drop(
    columns=[
        "user_id",
        "user_name",
        "review_id",
        "review_title",
        "review_content"
    ]
)

df.head()
```

```python
df["category"] = (
    df["category"]
    .str.replace("|", " ", regex=False)
    .str.replace("&", " ", regex=False)
)

df["category"].head()
```

```python
df["combined_features"] = (
    df["product_name"] + " " +
    df["category"] + " " +
    df["about_product"]
)

df[["product_name", "combined_features"]].head()
```

```python
print("Before removing duplicates:", df.shape)

df = df.drop_duplicates(
    subset=["product_name"]
).reset_index(drop=True)

print("After removing duplicates:", df.shape)
```

```python
df.to_csv("../data/processed/amazon_cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```
